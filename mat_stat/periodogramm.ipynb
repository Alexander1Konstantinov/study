```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv("data/9.csv")
px.line(data, x='t', y='xi', title='Изначальные данные', color_discrete_sequence=['purple'])
```

```python
n = data.shape[0]
aj = np.array([2/n * sum([data['xi'][i]*np.cos(2*np.pi*(j/n)*(i+1)) for i in range(n)]) for j in range(1,n+1)])
bj = np.array([2/n * sum([data['xi'][i]*np.sin(2*np.pi*(j/n)*(i+1)) for i in range(n)]) for j in range(1,n+1)])
```

```python
Pj = aj**2 + bj**2
px.line(x=range(n), y=Pj, color_discrete_sequence=['purple'], title='Периодограмма')
```

Получили, что наибольшую часть суммарной амплитуды составляют 3 и 517 гармоники

```python
print(f"Доля 3-й амплитуды c периодом {n/3}: {Pj[2]/np.sum(Pj):.3f}")
print(f"Доля 517-й амплитуды c периодом {n/517}: {Pj[516]/np.sum(Pj):.3f}")
```

```python
def seas(J, n, t):
    result=0
    for j in J:
        result += aj[j]*np.cos(2*np.pi*(j+1)/n*t) + bj[j]*np.sin(2*np.pi*(j+1)/n*t)
    return result
```

```python
J1 = [2,516] #выбранные гармоники
J2 = [2,13,505,516] # ещё две 

approx2 = [seas(J1, n, t) for t in np.arange(1,n+1,1)]
approx4 = [seas(J2, n, t) for t in np.arange(1,n+1,1)]
data['approx2'] = approx2
data['approx4'] = approx4
px.line(data, x='t', y=['xi','approx2', 'approx4'], color_discrete_sequence=['purple', 'red', 'blue'], title='Исходная выборочной траектория и найденные сезонной компоненты')
```
